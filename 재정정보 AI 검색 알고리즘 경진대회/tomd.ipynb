````python
import PyPDF2
import re

def pdf_to_markdown(pdf_path, output_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n\n"

    # 표 변환
    # text = convert_tables(text)

    # 박스 형태 내용 변환
    # text = convert_boxes(text)

    # 제목 변환
    # text = convert_titles(text)

    # Markdown 파일로 저장
    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(text)

def convert_tables(text):
    # 표 시작과 끝 감지
    table_pattern = r'\n(.+\n)+?-{3,}'
    tables = re.findall(table_pattern, text, re.MULTILINE)
    
    for table in tables:
        rows = table.strip().split('\n')
        markdown_table = "| " + " | ".join(rows[0].split()) + " |\n"
        markdown_table += "|" + "---|" * (len(rows[0].split()) - 1) + "---|\n"
        for row in rows[1:]:
            markdown_table += "| " + " | ".join(row.split()) + " |\n"
        text = text.replace(table, markdown_table)

    return text

def convert_boxes(text):
    # 박스 형태 내용 감지 및 변환
    box_pattern = r'□\s*(.+?)(?=\n□|\Z)'
    boxes = re.findall(box_pattern, text, re.DOTALL)
    
    for box in boxes:
        markdown_box = f"\n```\n{box.strip()}\n```\n"
        text = text.replace(f"□ {box}", markdown_box)

    return text

def convert_titles(text):
    # 제목 변환
    lines = text.split('\n')
    for i, line in enumerate(lines):
        if re.match(r'^[0-9]+\)\s*\S+', line):  # 숫자) 형태의 제목
            lines[i] = f"## {line}"
        elif re.match(r'^[0-9]+\.\s*\S+', line):  # 숫자. 형태의 제목
            lines[i] = f"### {line}"
    
    return '\n'.join(lines)

pdf_path = 'C:/Users/wjdwn/OneDrive/바탕 화면/Work/Dacon/재정정보 AI 검색 알고리즘 경진대회/data/open/test_source/산업통상자원부_에너지바우처.pdf'  # PDF 파일 경로
txt_path = 'C:/Users/wjdwn/OneDrive/바탕 화면/Work/Dacon/재정정보 AI 검색 알고리즘 경진대회/output2.txt'  # 저장할 TXT 파일 경로

# 사용 예
pdf_to_markdown(pdf_path, txt_path)
````
